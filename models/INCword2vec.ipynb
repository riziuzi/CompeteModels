```python
from gensim import utils
from collections import Counter
stopWord = ["the","of","and","in","is","to","are","it","from","by","as","on","with","or","this","they","its","at","an","into","between","which","have"] # all words appear more than 500 times (upto 4059)

class MyCorpus:
    """An iterator that yields sentences (lists of str)."""

    def __iter__(self):
        corpus_path = './geography.txt'
        for line in open(corpus_path, encoding='utf-8'):
            words = utils.simple_preprocess(line)
            # Remove stop words
            words = [word for word in words if word not in stopWord]

            yield words

def printFreq():
    corpus = MyCorpus()    
    flat_words = [word for words in corpus for word in words]

    word_frequencies = Counter(flat_words)

    sorted_words = sorted(word_frequencies.items(), key=lambda x: x[1], reverse=True)

    for word, frequency in sorted_words:
        print(f'{word}: {frequency}')

def printSentences():
    corpus = MyCorpus()
    for i in corpus:
        print(i)


printSentences()
```

```python
import gensim.models
sentences = MyCorpus()
model = gensim.models.Word2Vec(sentences=sentences)
print(f"Done!\nVocab size : {len(model.wv.index_to_key)}\n")
# for index, word in enumerate(model.wv.index_to_key):
#     if index == 10:
#         break
#     print(word)
```

```python
s = "upsc"
positive = s.lower().split(" ")
positive_filtered = [word for word in positive if word in model.wv.index_to_key]
if positive_filtered:
    similar_words = model.wv.most_similar(positive=positive_filtered, topn=20)
    print(similar_words)
else:
    print("No words in the input are present in the model's vocabulary.")
```
